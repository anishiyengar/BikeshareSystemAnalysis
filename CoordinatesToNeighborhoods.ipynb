```python
import csv
import json
import requests
import urllib.request
import matplotlib.pyplot as plt

# open the file in universal line ending mode 
with open('Hubway_Stations_as_of_July_2017.csv', 'rU') as infile:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(infile)
  data = {}
  for row in reader:
    for header, value in row.items():
      try:
        data[header].append(value)
      except KeyError:
        data[header] = [value]

# extract the variables you want
stations = data['Station']
latitudes = data['Latitude']
longitudes = data['Longitude']

def get_geonames(lat, lng, types):
    url = "https://maps.googleapis.com/maps/api/geocode/json?latlng="+lat+","+lng+"&key=AIzaSyDQkBBuv61MXdM11whbonG4yy9lc12h5ig"
    jsondata = json.load(urllib.request.urlopen(url))
    address_comps = jsondata['results'][0]['address_components']
    filter_method = lambda x: len(set(x['types']).intersection(types))
    return filter(filter_method, address_comps)

neighborhoods = []

for station in range(len(stations)):
    lat = latitudes[station]
    lng = longitudes[station]
    types = ['neighborhood', 'political']
    
    get_geonames(lat, lng, types)
    
    for geoname in get_geonames(lat, lng, types):
        common_types = set(geoname['types']).intersection(set(types))
        result = '{}'.format(geoname['long_name'])
        print(result)
        neighborhoods += [result]
        break

#print(neighborhoods)
len(neighborhoods)

station_freq = {}

for hubway in neighborhoods:
    if hubway not in station_freq:
        station_freq[hubway] = 1
    else:
        station_freq[hubway] += 1
        
print(station_freq)
        
plt.bar(list(station_freq.keys()), station_freq.values(), color='g')
#plt.title('Hubway Station per Boston Neighborhood')
#plt.xlabel('Number of Hubway Stations')
#plt.show('Boston Neighborhood')
plt.savefig('Neighborhood_Hubway_Frequency.png')
```
